```python
df_wnem4 = pd.read_csv('C:/Users/sdavedu/Desktop/XGBoost/WNEM4_2018_2019.csv', header = 0, index_col=False)
df_wnem4.columns = ['Date', 'WNEM_actual', 'WNEM_pred', 'WNEM_baseline']
df_wnem4.head()
```

```python
df_merge1 = df_dtlm4.merge(df_ltrm4, on = 'Date', how = 'inner')
df_merge1.shape
```

```python
pd.set_option('display.max_rows', 9999)
df_merge = df_merge1.merge(df_merge2, on = 'Date', how = 'inner')
df_merge.shape
```

```python
df_merge.head()
```

```python
df_merge.tail()
```

```python
df_merge['total_ice_cover_pred'] = df_merge[['DTLM_pred', 'LTRM_pred', 'WNEM_pred', 'SWPM_pred']].mean(axis = 1)
df_merge['total_ice_cover_actual'] = df_merge[['DTLM_actual', 'LTRM_actual', 'WNEM_actual', 'SWPM_actual']].mean(axis = 1)
df_merge['total_ice_cover_baseline'] = df_merge[['DTLM_baseline', 'LTRM_baseline', 'WNEM_baseline', 'SWPM_baseline']].mean(axis = 1)

df_merge.head()
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
```

```python
print(mean_absolute_error(df_merge['total_ice_cover_actual'], df_merge['total_ice_cover_pred']))
print(math.sqrt(mean_squared_error(df_merge['total_ice_cover_actual'], df_merge['total_ice_cover_pred'])))
```

```python
df_merge = df_merge.set_index('Date')
df_merge['month'] = pd.DatetimeIndex(df_merge.index).month
df_merge['day'] = pd.DatetimeIndex(df_merge.index).day
```

```python
###### Freezing time is November 1st to January 14th

df_freezing = df_merge[df_merge['month'].isin([11, 12, 1])]
df_freezing = df_freezing[~((df_freezing['month'] == 11) & (df_freezing['day'] < 1))]
df_freezing = df_freezing[~((df_freezing['month'] == 1) & (df_freezing['day'] > 14))]
```

```python
###### Stable time is January 15th to Mar 25th

df_stable = df_merge[df_merge['month'].isin([1, 2, 3, 4])]
df_stable = df_stable[~((df_stable['month'] == 1) & (df_stable['day'] < 15))]
df_stable = df_stable[~((df_stable['month'] == 4) & (df_stable['day'] > 25))]
```

```python
###### Melting time is Mar 26th to May 10th

df_melting = df_merge[df_merge['month'].isin([4, 5])]
df_melting = df_melting[~((df_melting['month'] == 4) & (df_melting['day'] < 26))]
df_melting = df_melting[~((df_melting['month'] == 5) & (df_melting['day'] > 10))]
```

```python
######## Printing Freezing MAE and RMSE ########
print(mean_absolute_error(df_freezing['total_ice_cover_actual'], df_freezing['total_ice_cover_pred']))
print(math.sqrt(mean_squared_error(df_freezing['total_ice_cover_actual'], df_freezing['total_ice_cover_pred'])))
```

```python
######## Printing stable MAE and RMSE ########
print(mean_absolute_error(df_stable['total_ice_cover_actual'], df_stable['total_ice_cover_pred']))
print(math.sqrt(mean_squared_error(df_stable['total_ice_cover_actual'], df_stable['total_ice_cover_pred'])))
```

```python
######## Printing melting MAE and RMSE ########
print(mean_absolute_error(df_melting['total_ice_cover_actual'], df_melting['total_ice_cover_pred']))
print(math.sqrt(mean_squared_error(df_melting['total_ice_cover_actual'], df_melting['total_ice_cover_pred'])))
```

```python
# Overlaying the 2018-2019 predictions on top of actual data for 2018-2019

fig1 = plt.figure(figsize=(15,8))
df_merge['total_ice_cover_actual'].plot(label = 'actual ice cover')
df_merge['total_ice_cover_pred'].plot(label = 'predicted ice cover')
df_merge['total_ice_cover_baseline'].plot(label = 'baseline ice cover')
plt.legend(loc="lower left")
plt.title('Actual vs. Predicted vs. Baseline for 2018-2019 for all stations using XGBoost ' , fontsize=15)
```

```python
# Draw the Ice-on and Ice-off plots


df_original = df_merge_icedur_a  #actual
df = df_merge_icedur_p           #predicted

xValue_original = []
yValue_original = []
xValue = []
yValue = []
tick_spacing = 5

for i in range(1,len(df_original['Year'])):
    if(int(df_original['max_ice_off_a'][i].strftime("%m")) < 11):
        xValue_original.append(df_original['Year'][i])
        xValue_original.append(df_original['Year'][i])
    else:
        xValue_original.append(df_original['Year'][i]+1)
        xValue_original.append(df_original['Year'][i]+1)

for i in range(1,len(df['Year'])):
    if(int(df['max_ice_off_p'][i].strftime("%m")) < 11):
        xValue.append(df['Year'][i])
        xValue.append(df['Year'][i])
    else:
        xValue.append(df['Year'][i]+1)
        xValue.append(df['Year'][i]+1)
        
    
for i in range(0,(len(df_original['max_ice_off_a'])-1)):
    if(int(df_original['min_ice_on_a'][i].strftime("%m")) >= 11):
        start_date = datetime(int(df_original['min_ice_on_a'][i].strftime("%Y")),11,1,0,0,0)
    yValue_original.append((df_original['min_ice_on_a'][i]-start_date).days)
    yValue_original.append((df_original['max_ice_off_a'][i+1]-start_date).days)

for i in range(0,(len(df['max_ice_off_p'])-1)):
    if(int(df['min_ice_on_p'][i].strftime("%m")) >= 11):
        start_date = datetime(int(df['min_ice_on_p'][i].strftime("%Y")),11,1,0,0,0)
    else:
        start_date = datetime(int(df['min_ice_on_p'][i].strftime("%Y"))-1,11,1,0,0,0)
    yValue.append((df['min_ice_on_p'][i]-start_date).days)
    yValue.append((df['max_ice_off_p'][i+1]-start_date).days)


      
plt.figure(figsize=(16,8))  

for i in range(0, int(len(xValue)/2)):

    plt.annotate(df['min_ice_on_p'][i].strftime("%Y-%m-%d"),(xValue[2*i],yValue[2*i]), xytext=(xValue[2*i]+0.25,yValue[2*i] -2), fontsize=8)

for i in range(0, int(len(xValue_original)/2)):
    if(i==0):
        line1 = plt.plot([xValue_original[2*i], xValue_original[2*i+1]],[yValue_original[2*i], yValue_original[2*i+1]],  color ='k', label = 'Original')
        
    line1 = plt.plot([xValue_original[2*i], xValue_original[2*i+1]],[yValue_original[2*i], yValue_original[2*i+1]],  color ='k')

plt.legend()
    #     plt.annotate(df['Ice Duration'][i], ((xValue[2*i]+xValue[2*i+1])/2, (yValue[2*i]+yValue[2*i+1])/2), ((xValue[2*i]+xValue[2*i+1])/2+0.1, (yValue[2*i]+yValue[2*i+1])/2))
    
for j in range(0, int(len(xValue)/2)):
    if(j == 0):
        line2 = plt.plot([xValue[2*j] + 0.3, xValue[2*j+1]+0.3],[yValue[2*j], yValue[2*j+1]], label = 'Predicted', color = 'r')
        
    line2 = plt.plot([xValue[2*j] + 0.3, xValue[2*j+1]+0.3],[yValue[2*j], yValue[2*j+1]], color = 'r')  

plt.legend(loc='upper right', fontsize = 12)




plt.title('Ice Season Duration Original vs Predicted for whole river - XGBoost ' , fontsize=20)

# plt.axis('off')
plt.yticks([])


frame = plt.gca().axes.get_xaxis().set_visible(False)

plt.savefig('Ice_duration_river_XGBoost.png')
```
